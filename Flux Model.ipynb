```python
%%writefile flux_model.jl
# Add packages
using Pkg
Pkg.add(["Flux", "Images", "FileIO", "NPZ", "Statistics", "BSON"])

using Flux
using Images
using FileIO
using NPZ
using Statistics
using BSON: @save

# Get command line arguments
img_path = ARGS[1]

# Load and preprocess the image
println("Loading and preprocessing image from: ", img_path)
img_tensor = npzread(img_path)
    
println("Original tensor shape: ", size(img_tensor))

# Ensure the tensor is in the right format for Flux
# Flux expects (height, width, channels, batch)
if length(size(img_tensor)) == 3
    # If tensor is (channels, height, width), permute to (height, width, channels)
    if size(img_tensor)[1] in [1, 3, 4] && size(img_tensor)[2] > 4 && size(img_tensor)[3] > 4
        img_tensor = permutedims(img_tensor, (2, 3, 1))
        println("Permuted tensor shape (height, width, channels): ", size(img_tensor))
    end
    
    # Add batch dimension
    img_tensor = reshape(img_tensor, (size(img_tensor)..., 1))
    println("Final tensor shape with batch (height, width, channels, batch): ", size(img_tensor))
end

# Convert to Float32
img_tensor = Float32.(img_tensor)

# Get dimensions for model creation
height, width, channels, batch = size(img_tensor)

# Calculate dimensions after max pooling twice
height_after_pool = height ÷ 4
width_after_pool = width ÷ 4

# Define a simple CNN model
model = Chain(
    # First convolutional layer
    Conv((3, 3), channels => 16, relu; pad=1),
    MaxPool((2, 2)),
    
    # Second convolutional layer
    Conv((3, 3), 16 => 32, relu; pad=1),
    MaxPool((2, 2)),
    
    # Flatten the output for the dense layer
    Flux.flatten,
    
    # Dense layers
    Dense(32 * height_after_pool * width_after_pool, 64, relu),
    Dense(64, 10),
    softmax
)

println("Model architecture:")
println(model)

# Perform forward pass
output = model(img_tensor)

println("Forward pass complete!")
println("Output shape: ", size(output))
println("Output values (first few):")
println(output[:, 1][1:5])

# Save the model
@save "flux_simple_cnn.bson" model
println("Model saved to flux_simple_cnn.bson")
```

```python
%%writefile visualize_filters.jl
using Pkg
Pkg.add(["Flux", "BSON", "Images", "FileIO"])

using Flux
using BSON: @load
using Images
using FileIO

# Load the model
@load "flux_simple_cnn.bson" model

println("Model loaded successfully")

# Extract filters from the first convolutional layer
first_layer = model[1]
weights = first_layer.weight

println("First layer filter shape: ", size(weights))

# Save filters as images
num_filters = size(weights, 4)
num_channels = size(weights, 3)

for i in 1:min(5, num_filters)
    # For each filter, create an image from the weights
    filter_img = Array{Float32}(undef, size(weights, 1), size(weights, 2), 3)
    
    for c in 1:min(3, num_channels)
        # Normalize the weights to the 0-1 range for visualization
        filter_weights = weights[:, :, c, i]
        min_w = minimum(filter_weights)
        max_w = maximum(filter_weights)
        norm_weights = (filter_weights .- min_w) ./ (max_w - min_w)
        
        # Assign to RGB channels
        filter_img[:, :, c] = norm_weights
    end
    
    # If there's only one channel, duplicate to all RGB channels
    if num_channels == 1
        filter_img[:, :, 2] = filter_img[:, :, 1]
        filter_img[:, :, 3] = filter_img[:, :, 1]
    end
    
    # Save the filter as an image
    save("filter_$(i).png", colorview(RGB, permutedims(filter_img, (3, 1, 2))))
    println("Saved filter_$(i).png")
end
```
